# RQ1: Build Graph 

```python
import networkx as nx
```

G = (V, E) where V is node(set of articles) and E is edge(hyperlink) 

```python
#Creating a Graph 
Graph = nx.Graph() #empty Graph 
```

```python
#loading the file and creating a dictionary with multiple values
d = {}

with open("wiki-topcats-reduced.txt") as data: 
    for line in data:                         
        items = line.split()
        key, values = int(items[0]), int(items[1])
        d.setdefault(key, []).append(values)
    data.close()
```

```python
d
```

```python
Graph = nx.Graph(d)
```

for undirected graph, (u,v) = (v,u)

```python
nx.is_directed(Graph) #checking to know if the Graph is directed or not 
```

The graph therefore is undirected. 

```python
V = nx.number_of_nodes(Graph)
V
```

```python
E = nx.number_of_edges(Graph)
E
```

Node degree or degree of the vertex is the number of edges connecting to it, with loops counted twice

```python
nx.info(Graph)
```

Average degree of nodes is 6.7699

```python
deg_V = 2 *abs(E)
deg_V
```

Graph Density 
This is defined as the ratio of the number of edges of a given graph, and the total number of edges the graph could have.

If the graph is complete, the maximum density is 1.
max. number of edges = 1/2 * |V| * (|V|-1)
Therefore Density, D = 2|E|/|V|(|V|-1)

NOTE: A dense graph is a graph in which the number of edges is close to the maximal number of edges.

```python
A = 2* abs(E)
B = abs(V)*(abs(V)-1)
```

```python
D = A/B
D
```
